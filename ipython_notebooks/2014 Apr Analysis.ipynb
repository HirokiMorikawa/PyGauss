```python
import pygauss.analysis as pg
%load_ext chemlab.ipython
%matplotlib inline

import os
home = os.path.expanduser("~")
file_path = os.path.join(home, 'Dropbox', 'Mres_Project', 'gaussian', 'final_files')
```

```python
mol = pg.Molecule(file_path, 
                  init_fname='CJS1_coro-emim_T-C4-C39-P_init.com',
                  opt_fname='CJS1_coro-emim_T-C4-C39-P_6-311g-d-p-_gd3bj_opt_i.log', 
                  freq_fname='CJS1_coro-emim_T-C4-C39-P_6-311g-d-p-_gd3bj_freq_i.log',
                  alignto=[1, 29, 13]) # align coronene surface with x-y surface
```

```python
print('Optimised:', mol.is_optimised())
mol.plot_optimisation_E()
```

```python
print('Conformer:', mol.is_conformer())
mol.plot_IRfreqs()
```

```python
mol.show_initial(ball_stick=True, rotations=[[0, 0, 0], [90, 0, 0], [0, 90, 0], [0, 0, 90]], zoom=2)
```

```python
mol.show_optimisation(active=True, opt_step=100)
```

```python
mol.show_highlight_atoms([range(1, 37), [39, 40, 45, 46], range(37, 56)], ball_stick=True, zoom=2., 
                    rotations=[[0, 0, 0], [90, 0, 0], [0, 90, 0], [0, 0, 90]])
```
